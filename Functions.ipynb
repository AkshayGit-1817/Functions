6. Explain the concept of generators in Python and how they are defined.
:-
   Generators are special functions that return a lazy iterator, producing values one at a time on demand.
  
   * Definition: They use the yield keyword instead of return.
   * State: They pause execution and save their state between values.
   * Resuming: The next value is requested using the next() function or a loop.

7. What are the advantages of using generators over regular functions?
:-  
   Generators offer significant performance and design benefits over regular functions that return lists.
   
   * Memory Efficiency: They do not store the entire sequence in memory.
   * Infinite Streams: They can model infinite data structures safely.Time * Efficiency: They start producing outputs immediately before processing the entire dataset.

```python
1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.
:-
   def sum_even_numbers(numbers):
    """
    Takes a list of numbers and returns the sum of all even numbers.
    """
    return sum(num for num in numbers if num % 2 == 0)

# Example Usage:
# my_list = [1, 2, 3, 4, 5, 6]
# print(sum_even_numbers(my_list))  # Output: 12 (2 + 4 + 6)

2. Create a Python function that accepts a string and returns the reverse of that string.
:-
  def reverse_string(text):
    """
    Accepts a string and returns its reversed version.
    """
    return text[::-1]

# Example Usage:
# my_str = "hello"
# print(reverse_string(my_str))  # Output: "olleh"

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.
:-
  def square_numbers(numbers):
    """
    Takes a list of integers and returns a new list containing their squares.
    """
    return [num ** 2 for num in numbers]

# Example Usage:
# my_list = [1, 2, 3, 4]
# print(square_numbers(my_list))  # Output: [1, 4, 9, 16]

4. Write a Python function that checks if a given number is prime or not from 1 to 200.
:-
  def is_prime(n):
    """
    Checks if a number n (optimized for the range 1 to 200) is prime.
    Returns True if prime, False otherwise.
    """
    if n <= 1:
        return False
    if n <= 3:
        return True  # 2 and 3 are prime
    if n % 2 == 0 or n % 3 == 0:
        return False

    # Check divisors up to the square root of n (max √200 ≈ 14.14)
    # Loop checks 5 and 11 (6k +/- 1 optimization)
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6

    return True

# Example Usage to check numbers from 1 to 200:
# primes_1_to_200 = [num for num in range(1, 201) if is_prime(num)]
# print(primes_1_to_200)

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.
:-
  class FibonacciIterator:
    """
    An iterator class that generates the Fibonacci sequence
    up to a specified number of terms.
    """
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration

        # Save the current value to return
        current = self.a
        # Compute the next numbers in the sequence
        self.a, self.b = self.b, self.a + self.b
        # Increment the counter
        self.count += 1

        return current

# Example Usage:
# fib_seq = FibonacciIterator(7)
# print(list(fib_seq))  # Output: [0, 1, 1, 2, 3, 5, 8]

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
:-
  def powers_of_two(max_exponent):
    """
    A generator function that yields powers of 2
    from 2^0 up to 2^max_exponent.
    """
    exponent = 0
    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1

# Example Usage:
# for value in powers_of_two(5):
#     print(value)  # Output: 1, 2, 4, 8, 16, 32

7. Implement a generator function that reads a file line by line and yields each line as a string.
:-
  def read_lines_lazy(file_path):
    """
    A generator function that reads a file line by line safely,
    yielding each line as a string without loading the whole file into memory.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.strip()

# Example Usage:
# for line in read_lines_lazy("large_data.txt"):
#     print(line)

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
:-
  def sort_tuples_by_second(tuple_list):
    """
    Sorts a list of tuples in-place (or returns a sorted copy)
    based on the second element of each tuple using a lambda function.
    """
    # Using sorted() to return a new sorted list
    return sorted(tuple_list, key=lambda item: item[1])

# Example Usage:
# data = [("apple", 3), ("banana", 1), ("cherry", 5)]
# print(sort_tuples_by_second(data))  # Output: [('banana', 1), ('apple', 3), ('cherry', 5)]

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
:-
   def celsius_to_fahrenheit(celsius_list):
    """
    Converts a list of Celsius temperatures to Fahrenheit using map() and lambda.
    Formula: (C * 9/5) + 32
    """
    # map() applies the lambda formula to every item in the celsius_list
    fahrenheit_iterator = map(lambda c: (c * 9/5) + 32, celsius_list)

    # Convert the resulting iterator back into a list
    return list(fahrenheit_iterator)

# Example Usage:
# temps_c = [0, 20, 37, 100]
# print(celsius_to_fahrenheit(temps_c))  # Output: [32.0, 68.0, 98.6, 212.0]

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
:-
  def remove_vowels(text):
    """
    Removes all vowels (both lowercase and uppercase) from a string
    using the filter() function.
    """
    vowels = "aeiouAEIOU"

    # filter() keeps characters where the lambda returns True (not in vowels)
    filtered_chars = filter(lambda char: char not in vowels, text)

    # Join the remaining characters back into a single string
    return "".join(filtered_chars)

# Example Usage:
# my_string = "Hello World, Python is Awesome!"
# print(remove_vowels(my_string))  # Output: "Hll Wrld, Pthn s wsm!"

Q 11. ~~~~~~~~

:-
  # Input data structure based on the image table layout
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Map processes each sublist:
# item[0] is Order Number, item[2] is Quantity, item[3] is Price per Item
processed_orders = list(map(
    lambda item: (
        item[0],
        (item[2] * item[3]) if (item[2] * item[3]) >= 100.00 else (item[2] * item[3]) + 10.00
    ),
    orders
))

# Print results rounded to 2 decimal places for presentation
for order_num, total_price in processed_orders:
    print(f"({order_num}, {total_price:.2f})")
```
